```python
#1.Email Validator (String Filtering + Loops)
üíº Real-world use: Used in signup forms or data cleaning tasks.
Task:
Check if a given email is valid ‚Äî must contain ‚Äú@‚Äù and end with ‚Äú.com‚Äù or ‚Äú.org‚Äù.
Example:
Input: hello@inceptionbd.com
Output: Valid Email
```

```python
# Input email
email = input("Enter your email: ")

# Initialize flags
has_at = False
valid_domain = False

# Check if '@' exists in email
for char in email:
    if char == '@':
        has_at = True
        break

# Check if email ends with '.com' or '.org'
for domain in ['.com', '.org']:
    if email.endswith(domain):
        valid_domain = True
        break

# Final validation
if has_at and valid_domain:
    print("Valid Email")
else:
    print("Invalid Email")
```

```python
#2. Count Word Frequency in Customer Reviews
üíº Real-world use: Analyzing reviews or comments in text analytics.
Task:
Given a review text, count how many times each word appears (ignore case).
Example:
Input: "This product is good and this service is excellent"
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}
```

```python
# Input review text
review = input("Enter customer review: ")

# Convert text to lowercase to ignore case
review_lower = review.lower()

# Split text into words
words = review_lower.split()

# Initialize an empty dictionary to store word counts
word_count = {}

# Loop through each word and count frequency
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Print the result
print(word_count)
```

```python
#3. Password Strength Checker
üíº Real-world use: Used in user authentication systems.
Task:
Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one
special character.
Example:
Input: "Hello@123"
Output: Strong Password
```

```python

password = input("Enter your password: ")

# Initialize flags
has_upper = False
has_lower = False
has_digit = False
has_special = False

# Define special characters
special_characters = "!@#$%^&*()-_+=<>?/|\\{}[]~`"

# Loop through each character in password
for char in password:
    if char.isupper():
        has_upper = True
    elif char.islower():
        has_lower = True
    elif char.isdigit():
        has_digit = True
    elif char in special_characters:
        has_special = True

# Check all conditions
if has_upper and has_lower and has_digit and has_special:
    print("Strong Password")
else:
    print("Weak Password")
```

```python
#4. Username Generator
üíº Real-world use: Automatically create usernames for users.
Task:
Take user‚Äôs full name and generate a short username.
Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).
Example:
Input: "Mamun Malitha"
Output: mamitha57
```

```python

import random

# Input full name
full_name = input("Enter your full name (first last): ")

# Split the name into first and last
name_parts = full_name.split()

if len(name_parts) < 2:
    print("Please enter both first and last name.")
else:
    first_name = name_parts[0]
    last_name = name_parts[1]

    # Get first 3 letters of first name
    first_part = ""
    count = 0
    for char in first_name:
        if count < 3:
            first_part += char.lower()
            count += 1

    # Get last 3 letters of last name
    last_part = ""
    count = 0
    for char in reversed(last_name):
        if count < 3:
            last_part = char.lower() + last_part
            count += 1

    # Generate a random 2-digit number using loop
    number = ""
    for _ in range(2):
        number += str(random.randint(0, 9))

    # Combine to form username
    username = first_part + last_part + number

    print("Generated username:", username)
```

```python
#5. SMS Word Counter
üíº Real-world use: Used in messaging apps or social media post limit check.
Task:
Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).
Example:
Input: "This is a demo message."
Output: Words = 5, Characters = 23, Status = Within Limit
```

```python
# Input message
message = input("Enter your message: ")

# Count characters
char_count = 0
for char in message:
    char_count += 1

# Count words
words = message.split()
word_count = 0
for word in words:
    word_count += 1

# Check SMS limit
if char_count > 160:
    status = "Exceeded Limit"
else:
    status = "Within Limit"

# Display results
print("Words =", word_count)
print("Characters =", char_count)
print("Status =", status)
```

```python
#6. Remove Duplicates from Contact List
üíº Real-world use: Data cleaning in CRM or user databases.
Task:
Given a list of names (comma separated), remove duplicates.
Example:
Input: "Rahim, Karim, Rahim, Sultana"
Output: "Rahim, Karim, Sultana"
```

```python
# Input names as comma-separated string
names_input = input("Enter names separated by commas: ")

# Split names into a list and remove leading/trailing spaces
names_list = []
for name in names_input.split(","):
    names_list.append(name.strip())

# Initialize a list to store unique names
unique_names = []

# Loop through names and add to unique list if not already present
for name in names_list:
    if name not in unique_names:
        unique_names.append(name)

# Join the unique names back into a string
result = ", ".join(unique_names)

print("Unique Contact List:", result)
```

```python
#7. Detect Suspicious Words in Text (Moderation)
üíº Real-world use: Used in chat moderation or email filtering.
Task:
If any ‚Äúbanned‚Äù words like ['spam', 'scam', 'fake'] are found in a text, print ‚ÄúWarning‚Äù.
Example:
Input: "This is a scam offer"
Output: Warning: Suspicious content detected
```

```python
# Input text
text = input("Enter the text: ")

# Convert text to lowercase for case-insensitive check
text_lower = text.lower()

# List of banned words
banned_words = ['spam', 'scam', 'fake']

# Flag to detect suspicious content
warning_flag = False

# Loop through banned words
for word in banned_words:
    if word in text_lower:
        warning_flag = True
        break

# Output result
if warning_flag:
    print("Warning: Suspicious content detected")
else:
    print("Text is clean")
```

```python
#8. Invoice Code Formatter
üíº Real-world use: Formatting IDs or codes in finance or ERP systems.
Task:
Given an invoice number like 23, generate code in the format INV00023.
Example:
Input: 23
Output: INV00023
```

```python
# Input invoice number
invoice_number = input("Enter invoice number: ")

# Convert to integer
num = int(invoice_number)

# Initialize formatted number as string
formatted_number = ""

# Count how many zeros needed to make length 5
num_str = str(num)
zeros_needed = 5 - len(num_str)

# Add leading zeros using a loop
for _ in range(zeros_needed):
    formatted_number += "0"

# Add the invoice number
formatted_number += num_str

# Combine with prefix
invoice_code = "INV" + formatted_number

print("Formatted Invoice Code:", invoice_code)
```

```python
#9. Count Hashtags and Mentions (Social Media Analytics)
üíº Real-world use: Used in analyzing social media posts.
Task:
Count how many hashtags (#) and mentions (@) appear in a post.
Example:
Input: "Check out #AIWorkshop and follow @inceptionbd"
Output: Hashtags = 1, Mentions = 1
```

```python
# Input social media post
post = input("Enter your post: ")

# Initialize counters
hashtag_count = 0
mention_count = 0

# Loop through each character in the post
for char in post:
    if char == '#':
        hashtag_count += 1
    elif char == '@':
        mention_count += 1

# Display results
print("Hashtags =", hashtag_count)
print("Mentions =", mention_count)
```

```python
#10. Word Reversal for Data Encryption
üíº Real-world use: Simple text obfuscation or playful encoding.
Task:
Reverse each word in a sentence while keeping word order the same.
Example:
Input: "AI is fun"
Output: "IA si nuf"
```

```python
# Input sentence
sentence = input("Enter a sentence: ")

# Split the sentence into words
words = sentence.split()

# Initialize list to store reversed words
reversed_words = []

# Loop through each word
for word in words:
    reversed_word = ""
    # Reverse the word using a loop
    for char in word:
        reversed_word = char + reversed_word
    reversed_words.append(reversed_word)

# Join the reversed words to form the final sentence
encrypted_sentence = " ".join(reversed_words)

print("Encrypted Sentence:", encrypted_sentence)
```
